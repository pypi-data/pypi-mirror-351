```python
from burr.core import action, ApplicationContext, State, ApplicationBuilder
```

```python
@action(reads=[], writes=[])
def with_context(state: State, __context: ApplicationContext) -> State:
    print(__context)
    return state
```

```python
app = (
    ApplicationBuilder()
    .with_actions(with_context)
    .with_entrypoint("with_context")
    .with_tracker(project="milan_bug")
    .with_transitions(("with_context", "with_context"))
    .build()
)
```

```python
app
```

```python
app.run(halt_after=["with_context"])
```
