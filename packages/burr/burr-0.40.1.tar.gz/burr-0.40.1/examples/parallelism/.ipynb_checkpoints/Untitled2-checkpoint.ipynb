```python
from burr.core import State, action, ApplicationContext, ApplicationBuilder
from burr.core.parallelism import MapActions, RunnableGraph
from burr.core.graph import GraphBuilder

from typing import Generator, Dict, Any, List
import random
import time

def sleep_random(max_sleep: float=1):
    return time.sleep(random.random()*max_sleep)
    
```

```python
@action(reads=[], writes=["num"])
def process_input(state: State, num: int) -> State:
    return state.update(num=state["num"])
    
@action(reads=["num"], writes=["num"])
def increment(state: State, increment_by: int) -> State:
    sleep_random()
    return state.update(output=state["input"] + increment_by)

@action(reads=["num"], writes=["num"])
def multiply(state: State, multiply_by: int) -> State:
    sleep_random()
    return state.update(output=state["input"] * multiply_by)


@action(reads=["num"], writes=["numbers"])
def explode(state: State, chunks: int) -> State:
    selections  = [random.randint(0, state["num"]) for _ in range(chunks)]


@action(reads=["nums"], writes=["num"])
def sum(state: State) -> State:
    return state.update(final_result=sum(state["nums"]))

class Recursive(MapActions):
    def __init__(self, recursion_depth: int, max_recursion_depth: int):
        super(Recursive, self).__init__()
        self.recursion_depth = recursion_depth
        self.max_recursion_depth = max_recursion_depth

    def actions(state: State, context: ApplicationContext, inputs: Dict[str, Any]) -> Generator[RunnableGraph, None, None]:
        for input_num in state["nums"]:
            yield create_graph(
                recursion_depth=self.recursion_depth,
                max_recursion_depth=self.max_recursion_depth,
                input_num=state["num"]
            )    

    def state(state: State) -> RunnableGraph:
        return State() # empty, we don't need it as it gets bound...

    def reads() -> List[str]:
        return ["numbers"]

    def writes() -> List[str]:
        return ["numbers"]

    def reduce(state: State, states: Generator[State, None, None]) -> State:
        numbers = []
        for state_ in states:
            numbers.append(state["num"])
        return state.update(numbers=numbers)
```

```python
def create_graph(recursion_depth: int, max_recursion_depth: int, input_num: int=None):
    should_recur = recursion_depth < max_recursion_depth
    num_recurrences = random.randint(1,3)
    if should_recur:
        return (
            GraphBuilder()
            .with_actions(
                process_input.bind(num=input_num) if input_num is not None else process_input, 
                increment, 
                multiply, 
                explode, 
                sum,
                recur=Recursive(recursion_depth=recursion_depth+1, max_recursion_depth=max_recursion_depth)
            )
            .with_transitions(
                ("process_input", "increment"),
                ("increment", "multiply"),
                ("multiply", "explode"),
                ("explode", "recur"),
                ("recur", "sum")
            )
            .build()
        )
    else:
        return (
            GraphBuilder()
            .with_actions(
                process_input.bind(num=input_num), 
                increment, 
                multiply, 
            )
            .with_transitions(
                ("process_input", "increment"),
                ("increment", "multiply")       
            )
            .build()
        )
```

```python
app = (
    ApplicationBuilder()
    .with_graph(create_graph(recursion_depth=0, max_recursion_depth=3, input_num=None))
    .with_state(State())
    .with_tracker(project="test_recursion")
    .with_entrypoint("process_input")
    .build()
)
```
