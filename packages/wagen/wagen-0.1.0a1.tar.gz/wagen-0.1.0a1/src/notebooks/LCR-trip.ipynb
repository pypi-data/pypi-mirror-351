```python
import wapgen.caltopo_io as cti
import wapgen.timeline as tl
import pandas as pd
from pprint import pp
import datetime as dt
```

```python
cfg = dict(
    start_time=dt.datetime.fromisoformat('2025-09-23'),
    morning_daylight=1.5,
    evening_daylight=2.0,
    include_dt=False)
```

# To Little Colorado River (LCR). Rafting option.
Optional 8 mile packraft from LCR to above Tanner Rapid.
The Colorado river in GC flows and average speed of about 4 mph.

```python
raft_df, trip = cti.travelplan_multi_csv(csvfname='~/Downloads/travelplan.GC-LCR-raft.csv')
display(trip)
display(raft_df.round(1))
```

## Timeline

```python
times_df, meta = tl.timeline(raft_df, trip, **cfg)

print(trip)
print(meta)
with pd.option_context("display.max_columns", None):
    display(times_df.round(1))
```

# To Little Colorado River (LCR). NO RAFT option.

```python
df, trip, meta = tl.trip_timeline(csvfname='~/Downloads/travelplan.GC-LCR-noraft.csv', **cfg)
print(trip)
display(meta)
display(df.round(1))
```
